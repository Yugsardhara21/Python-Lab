### 12) WAP to accept date in dd mm yyyy format and display in multiple formats.
#### Input : 01 12 2025
#### Output :
##### 01/12/2025
##### 2025-12-01

```python
a,b,c = input("Input: ").split()
a = int(a)
b = int(b)
c = int(c)

print(f"{a}",f"{b}",f"{c}",sep="/")
print(f"{c}",f"{b}",f"{a}",sep="-")
```
