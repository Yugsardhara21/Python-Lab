### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter Number..."))
temp = n
rev = 0

while(n!=0):
    rev = (rev*10) + int(n%10)
    n = n//10

if(rev == temp):
    print("Palindrome")
else:
    print("Not Palindrome")
```
