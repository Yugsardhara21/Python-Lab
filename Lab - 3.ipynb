### 05) WAP to check whether the given year is leap year or not.

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
a = int(input("Enter Number"))
b = int(input("Enter Number"))
c = input("Enter Operator")

match c:
    case '+':
        print(f"Sum: {a+b}")
    case '-':
        print(f"Sum: {a-b}")
    case '*':
        print(f"Sum: {a*b}")
    case '/':
        print(f"Sum: {a/b}")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
a = int(input("Enter Mark 1"))
b = int(input("Enter Mark 2"))
c = int(input("Enter Mark 3"))
d = int(input("Enter Mark 4"))
e = int(input("Enter Mark 5"))

sum = a + b + c + d + e
p = (sum/5)
print(f"Percentage: {p}")

if(p  > 70):
    print(f"Distinction")
elif(p  > 60):
    print(f"First Class")
elif(p  > 45):
    print(f"Second Class")
elif(p  > 35):
    print(f"Pass")
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
units = int(input("Enter electricity units used: "))

bill = 0

if units <= 50:
    bill = units * 2.60

elif units <= 100:
    bill = (50 * 2.60) + (units - 50) * 3.25

elif units <= 200:
    bill = (50 * 2.60) + (50 * 3.25) + (units - 100) * 5.26

else:
    bill = (50 * 2.60) + (50 * 3.25) + (100 * 5.26) + (units - 200) * 8.45

print(f"Total Electricity Bill = Rs. {bill:.2f}")
    
```
