### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
str = input("Enter String")
uc=0
lc=0
dc=0
sc=0

for i in range(len(str)):
    if(str[i].isupper()):
        uc += 1
    elif(str[i].islower()):
        lc += 1
    elif(str[i].isdigit()):
        dc += 1
    elif(str[i].swapcase()):
        sc += 1

print(f"Input:{str} -> Uppercase : {uc}, Lowercase : {lc}, Digit : {dc}, Special : {sc}")    
```

### 2) Reverse Each Word

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
a = input("Enter String")
count = 1
b = ""

for i in range(1,len(a)):
    if(a[i]==a[i-1]):
        count += 1 
    else:
        b = b + a[i-1] + str(count)
        count = 1
        
b = b + a[-1] + str(count)
print(b)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
a = input("Enter String")
b = ""

for i in range(len(a)):
    if(a[i].isupper()):
        b += a[i].lower()
    else:
        b += a[i].upper()
print(b)
```

### 11) Username Validation
#### valid username rules:
1. only alphabets and digits
2. does not start with a digit
3. length ≥ 6

- `User123` → Valid
- `123User` → Invalid
- `Us@12` → Invalid

```python
a = input("Enter String: ")

if a[0].isdigit():
    print("Invalid")

elif len(a) < 6:
    print("Invalid")

else:
    for i in range(len(a)):
        if not a[i].isalnum():
            print("Invalid")
            break
    else:
        print("Valid")
```
